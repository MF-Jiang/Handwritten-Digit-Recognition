```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml
```

```python
# Check if the data file already exists
try:
    data = np.load('mnist_data.npz', allow_pickle=True)
    X_small = data['X_small']
    y_small = data['y_small']
#     print(y_small)
except FileNotFoundError:
    # If the data file does not exist, download and save the data
    X, y = fetch_openml('mnist_784', data_home='./', return_X_y=True)
    X = X / 255.
    X_small = X[:1000]
    y_small = y[:1000]
    # Save the data
    np.savez('mnist_data.npz', X_small=X_small, y_small=y_small)
```

```python
print(X_small.shape)

# print(len(np.nonzero(y_small.values.astype('int') == 0)[0]))
print(len(np.nonzero(y_small.astype('int') == 0)[0]))
```

```python
plt.figure()
index = np.nonzero(y_small.astype('int') == 2)[0]

for i in range(len(index)):
    plt.subplot(11, 11, i+1)
    plt.imshow(X_small[index[i]].reshape(28, 28), cmap='gray')
    plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Normalization -- x = (x-mean(x)) / std(x)
X_small_norm = np.zeros(X_small.shape)
for i in range(X_small.shape[0]):
    X_small_norm[i, :] = (X_small[i] - np.mean(X_small[i])) / np.std(X_small[i])
```

```python
# np.savetxt('array.txt', X_small_norm)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_small_norm, y_small, test_size=0.2, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = (y_pred == y_test).mean()
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
indices_of_2 = np.where(y_pred == '2')[0]

plt.figure()
plt.figure(figsize=(12, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[indices_of_2[i]].reshape(28, 28), cmap='gray')
    plt.title(f'Predicted: {y_pred[indices_of_2[i]]}')
    plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Find indices of misclassified samples
misclassified_indices = np.where(y_pred != y_test)[0]

# Create a figure for plotting
plt.figure()
plt.figure(figsize=(12, 5))
# Plot the first 10 correctly classified samples
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.title(f'Predicted: {y_pred[i]}')
    plt.xticks([]), plt.yticks([])
plt.show()
    
plt.figure()
plt.figure(figsize=(12, 5))
# Plot the first 10 misclassified samples
for i in range(10):
    index = misclassified_indices[i]
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[index].reshape(28, 28), cmap='gray')
    true_label = y_test[index]
    plt.title(f'Predicted: {y_pred[index]}, Actual: {true_label}')
    plt.xticks([]), plt.yticks([])

plt.show()
```
